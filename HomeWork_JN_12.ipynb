```python
import os
import matplotlib.pyplot as plt

root = os.getcwd()
filename=os.path.join(root, 'company_sales_data.csv')
filename
```

 - *данные о продажах (все)

```python
import pandas as pd
data=pd.read_csv(filename)
data.head()
```

```python
import calendar
month = list(calendar.month_abbr)
month.pop(0)
month
```

 **1. Прочитайте все данные о продажах продукта (*total_units*) и покажите их с помощью многострочного графика**

```python
facecream=data['facecream'].tolist() # данные о продажах facecream
facecream
```

```python
facewash=data['facewash'].tolist() # данные о продажах facewash
facewash
```

```python
toothpaste=data['toothpaste'].tolist() # данные о продажах toothpaste
toothpaste
```

```python
bathingsoap=data['bathingsoap'].tolist() # данные о продажах bathingsoap
bathingsoap
```

```python
shampoo=data['shampoo'].tolist() # данные о продажах shampoo
shampoo
```

```python
moisturizer=data['moisturizer'].tolist() # данные о продажах moisturizer
moisturizer
```

```python
plt.figure(figsize=(15,4))

plt.plot(month,facecream,label='facecream')
plt.plot(month,facewash,label='facewash')
plt.plot(month,toothpaste,label='toothpaste')
plt.plot(month,bathingsoap,label='bathingsoap')
plt.plot(month,shampoo,label='shampoo')
plt.plot(month,moisturizer,label='moisturizer')

plt.title('Year Sales units', fontsize=20, color='blue')
plt.xlabel('Month', fontsize=10, color='blue')
plt.ylabel('Month Unit', fontsize=10, color='blue')
plt.xticks(month)

plt.legend()
plt.show()
```

 **2. Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.**

 **3. Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы**

```python
list_sales = [sum(facecream), sum(facewash), sum(toothpaste), sum(bathingsoap), sum(shampoo), sum(moisturizer)]
list_sales
```

```python
list_product = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
list_product
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()
ax.pie(list_sales, labels=list_product)
wedges, texts, autotexts = ax.pie(list_sales, labels=list_product, autopct='%1.2f%%')
ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```
